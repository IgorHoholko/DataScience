### Preprocessing data for training
Create two data sets for model's trainig and fill our missing values with zeros and mean values (column) to check which one is better.

```python
# data for training
X = data_train.drop('Target', axis=1)
y = data_train['Target']

# data we have to predict Target values
X_pred = data_predict.drop('Target', axis=1)


X_zeros = X.fillna(0)
means = X[columns_nan].mean()
X_mean = X[columns_nan].fillna(means)
```

```python
from sklearn.model_selection import train_test_split

(X_train_zeros, X_test_zeros, y_train, y_test) = train_test_split(X_zeros, y,
                                                                 test_size=0.3, 
                                                                 random_state=0)

(X_train_mean, X_test_mean) = train_test_split(X_mean, test_size=0.3,  random_state=0)
```

```python
# check the distribution of the Target labels
plt.scatter(range(len(y_train)), y_train)
```

# Selecting the model

```python
from sklearn.metrics import roc_auc_score, roc_curve

def get_info(model, X, labels, pos_label = 1, thread = .5, draw = True):
    
    score = roc_auc_score(labels, model.predict_proba(X)[:, 1])
    FPR, TPR, thresholds  = roc_curve(labels, model.predict_proba(X)[:, 1], pos_label=pos_label)

    accuracy = sum( (model.predict_proba(X)[:, 1] >=thread) * 1 ==  labels) / len(y_test) 

    if draw:
        fig, axs = plt.subplots(nrows=1, ncols=3, sharex=True)
        fig.set_figheight(5)
        fig.set_figwidth(15)
        
        ax = axs[0]
        ax.plot(FPR, TPR, c='r')
        ax.set_title("ROC curve,\n AUC : {}".format(score))
        ax.set_xlabel('FPR')
        ax.set_ylabel('TPR')
        
        ax = axs[1]
        ax.plot(np.linspace(0, 1, len(FPR)), FPR)
        
        ax.set_ylabel('FPR')
        ax.set_title('FPR')
        
        ax = axs[2]
        ax.plot(np.linspace(0, 1, len(TPR)), TPR)
        ax.set_ylabel('TPR')
        ax.set_title('TPR')
        
        plt.show()
        
        print('accuracy: ', accuracy)
    
    return score
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10]}
cv = 3

estimator = LogisticRegression(penalty = 'l2')
optimizer_zeros = GridSearchCV(estimator, param_grid, cv = cv)
optimizer_zeros.fit(X_train_zeros, y_train)

estimator2 = LogisticRegression(penalty = 'l2')
optimizer_means = GridSearchCV(estimator2, param_grid, cv = cv)
optimizer_means.fit(X_train_mean, y_train)
print("Done!")
```

```python
_ = get_info(optimizer_zeros.best_estimator_, X_test_zeros, y_test)
```

```python
_ = get_info(optimizer_means.best_estimator_, X_test_mean, y_test)
```

    We can see that a model whose missing values are filled with zeros works better.

# Feature selection

```python
from sklearn.feature_selection import RFE
```

```python
accuracy = []

for i in range(2, 20):
    
    estimator = LogisticRegression(penalty = 'l2')
    rfe = RFE(estimator, i)

    fit = rfe.fit(X_train_zeros, y_train)

    columns_drop = X_train_zeros.columns[~fit.support_]
    X_selected_train = X_train_zeros.drop(columns_drop , axis=1)
    X_selected_test = X_test_zeros.drop(columns_drop , axis=1)
    
    model = xgboost.XGBClassifier(n_estimators=1000, max_depth=3)
    model.fit(X_selected_train, y_train)
    
    sc = get_info(model, X_selected_test, y_test, draw= False)
    accuracy.append([i, sc])
```

```python
accuracy = np.array(accuracy)
plt.plot(accuracy[:, 0], accuracy[:, 1])
```

    Best size = 8

### Define the function for predictins

```python
def predict(model, data, columns_drop, threshold = .5):
    if "Target" in data.columns:
        X = data.drop("Target", axis=1)
    
    if hasattr(columns_drop, "__len__"):
        X = X.drop(columns_drop , axis=1)
        labels = (model.predict_proba(X)[:, 1] >= threshold) * 1
        X.fillna(0, inplace=True)
    else:
        return None
    
    data.Target = labels
    data.to_excel('predicted_threshold_{}.xlsx'.format(threshold))
    return data
```
