```python
sn.pairplot(df)
```

```python
X = df.drop(labels='y', axis=1).values
y = df['y'].values
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=0)
```

# 1) Linear Regression 

```python
linear_regressor = LinearRegression()

linear_regressor.fit(X_train, y_train)

y_pred_linear = linear_regressor.predict(X_test)
```

```python
metrics.r2_score(y_test, linear_regressor.predict(X_test))
```

# 3) xgboost

```python
import xgboost
```

```python
XGB = xgboost.XGBRegressor(n_estimators=500, max_depth=5)

XGB.fit(X_train, y_train)

y_xgb_pred = XGB.predict(X_test)

print('MAE: ', metrics.mean_absolute_error(y_test, y_xgb_pred))
print('Determination: ', metrics.r2_score(y_test, y_xgb_pred))
```

# Neural Network

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(40, input_shape = (4,), activation='relu'),
    Dense(20, activation='relu'),
    Dense(10, activation='relu'),
    Dense(4, activation='relu'),
    Dense(1)

])

model.summary()

model.compile(Adam(lr = .001), loss = 'mean_squared_error',  metrics=['mae'])
```

To train the model, I changed the batch_size from 10 to 1000 and the number of epochs from 100 to 1000 to make more generalizations on the sample


```python
%%time
model.fit(X_train, y_train, batch_size=1000, epochs=1, shuffle=True)
```

```python
metrics.mean_absolute_error(y_test, model.predict(X_test))
```

```python
y_nn = model.predict(X_test)
```
